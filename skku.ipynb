```python
import asyncio
import nest_asyncio
from crawl4ai import AsyncWebCrawler
from bs4 import BeautifulSoup
import json

# Jupyter Notebook에서 실행할 수 있도록 설정
nest_asyncio.apply()

async def crawl_notices():
    async with AsyncWebCrawler() as crawler:
        result = await crawler.arun(
            url="https://www.skku.edu/skku/campus/skk_comm/notice01.do",
        )

    # HTML 파싱
    soup = BeautifulSoup(result.html, "html.parser")

    # 공지사항 데이터 저장 리스트
    notices = []

    # 공지사항 제목 및 URL 가져오기
    for item in soup.select(".board-list-content-title a"):
        title = item.text.strip()
        url = item["href"]

        # 상대경로를 절대경로로 변환
        if not url.startswith("http"):
            url = "https://www.skku.edu" + url

        notices.append({"title": title, "url": url})

    # JSON 파일로 저장
    with open("skku_notices.json", "w", encoding="utf-8") as f:
        json.dump(notices, f, ensure_ascii=False, indent=4)

    return notices

# 크롤링 실행
notices_data = await crawl_notices()

# ✅ 크롤링된 공지사항 미리보기
for notice in notices_data[:5]:
    print(f"🔹 {notice['title']}\n👉 {notice['url']}\n")
```

```python
import json

# 저장된 공지사항 JSON 불러오기
with open("skku_notices.json", "r", encoding="utf-8") as f:
    notices = json.load(f)

def search_notices(query):
    """사용자의 질문(query)에 맞는 공지사항을 검색하는 함수"""
    query = query.lower()
    results = [n for n in notices if query in n["title"].lower()]

    if results:
        print(f"\n🔍 '{query}' 관련 공지사항 {len(results)}개 찾았어!\n")
        for notice in results[:5]:  # 최대 5개까지만 출력
            print(f"📢 {notice['title']}\n👉 {notice['url']}\n")
    else:
        print("\n❌ 관련된 공지사항을 찾지 못했어. 다른 키워드로 검색해볼래?\n")

# 챗봇 실행
while True:
    query = input("\n🗣 찾고 싶은 공지사항 키워드를 입력하세요 (종료하려면 'exit' 입력): ")
    if query.lower() == "exit":
        print("\n👋 챗봇을 종료합니다.")
        break
    search_notices(query)
```
